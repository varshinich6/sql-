```
import json
import os
import random
import matplotlib.pyplot as plt

class User:
    def __init__(self, username, role):
        self.username = username
        self.role = role

class QuestionBank:
    def __init__(self):
        self.questions = {
            'Python': [
                ("What is the output of print(2 ** 3)?", 'a', 'a) 8', 'b) 6', 'c) 9', 'd) 4'),
                ("Which of the following is a mutable data type?", 'b', 'a) Tuple', 'b) List', 'c) String', 'd) Dictionary'),
                ("What keyword is used to define a function in Python?", 'c', 'a) function', 'b) defn', 'c) def', 'd) func'),
                ("What is the output of print(type([]))?", 'a', 'a) <class list>', 'b) <class dict>', 'c) <class set>', 'd) <class tuple>'),
                ("Which of the following is not a built-in data type in Python?", 'd', 'a) List', 'b) Dictionary', 'c) Set', 'd) Array'),
                ("How do you start a comment in Python?", 'b', 'a) /* comment */', 'b) # comment', 'c) // comment', 'd) <!-- comment -->'),
                ("What is the correct file extension for Python files?", 'c', 'a) .pyth', 'b) .pt', 'c) .py', 'd) .pyn'),
                ("Which of the following is used to create a set in Python?", 'a', 'a) {1, 2, 3}', 'b) [1, 2, 3]', 'c) (1, 2, 3)', 'd) <1, 2, 3>'),
                ("What is the keyword used to handle exceptions in Python?", 'b', 'a) catch', 'b) except', 'c) handle', 'd) try'),
                ("Which of the following is a correct syntax to return a value from a function?", 'c', 'a) return value', 'b) yield value', 'c) return (value)', 'd) return: value'),
            ],
            'Java': [
                ("What is the default value of an int variable in Java?", 'a', 'a) 0', 'b) null', 'c) undefined', 'd) 1'),
                ("Which of the following is not a Java keyword?", 'b', 'a) static', 'b) friend', 'c) void', 'd) public'),
                ("What is the superclass of all classes in Java?", 'c', 'a) Object', 'b) Class', 'c) Object', 'd) BaseClass'),
                ("Which of the following is used to create an object in Java?", 'd', 'a) new Object()', 'b) Object()', 'c) new()', 'd) new ClassName()'),
                ("What is the size of an int in Java?", 'a', 'a) 4 bytes', 'b) 2 bytes', 'c) 8 bytes', 'd) 1 byte'),
                ("Which of the following is a valid declaration of a char?", 'c', 'a) char c = 1;', 'b) char c = \"c\";', 'c) char c = \'c\';', 'd) char c = \'cc\';'),
                ("What is the output of System.out.println(1 + 2 + '3');", 'b', 'a) 6', 'b) 33', 'c) 12', 'd) 123'),
                ("Which of the following is a valid way to declare an array in Java?", 'a', 'a) int[] arr;', 'b) int arr[];', 'c) both a and b', 'd) None of the above'),
                ("What is the keyword used to inherit a class in Java?", 'c', 'a) extends', 'b) inherit', 'c) extends', 'd) implements'),
                ("Which of the following is a correct way to start a thread in Java?", 'd', 'a) new Thread().start();', 'b) Thread.start();', 'c) start Thread;', 'd) new Thread().start();'),
            ],
            'C#': [
                ("Which of the following is a value type in C#?", 'a', 'a) int', 'b) string', 'c ) object', 'd) dynamic'),
                ("What is the output of Console.WriteLine(5 + 5);", 'a', 'a) 10', 'b) 55', 'c) 5', 'd) 0'),
                ("Which of the following is used to define a class in C#?", 'c', 'a) class MyClass {}', 'b) MyClass class {}', 'c) class MyClass {}', 'd) define MyClass {}'),
                ("What is the default access modifier for class members in C#?", 'b', 'a) public', 'b) private', 'c) protected', 'd) internal'),
                ("Which of the following is a correct way to declare an array in C#?", 'c', 'a) int arr[];', 'b) int[] arr;', 'c) both a and b', 'd) None of the above'),
                ("What is the keyword used to handle exceptions in C#?", 'b', 'a) catch', 'b) try', 'c) handle', 'd) except'),
                ("Which of the following is a correct way to declare a string in C#?", 'c', 'a) string s = \"Hello\";', 'b) string s = \'Hello\';', 'c) both a and b', 'd) None of the above'),
                ("What is the output of Console.WriteLine(1 + 2 + '3');", 'b', 'a) 6', 'b) 33', 'c) 12', 'd) 123'),
                ("Which of the following is a correct way to start a thread in C#?", 'd', 'a) new Thread().Start();', 'b) Thread.Start();', 'c) start Thread;', 'd) new Thread().Start();'),
                ("What is the keyword used to define an interface in C#?", 'c', 'a) interface MyInterface {}', 'b) MyInterface interface {}', 'c) interface MyInterface {}', 'd) define MyInterface {}'),
            ]
        }

class AutomatedQuestionBuilderApp:
    def __init__(self):
        self.users = {}
        self.current_user = None
        self.question_bank = QuestionBank()

    def register_user(self, username, role):
        self.current_user = User(username, role)
        self.users[username] = self.current_user

    def self_assessment(self):
        if self.current_user.role != 'Employee':
            print("Access denied. Only employees can take the self-assessment.")
            return
        topic = input("Choose a topic (Python, Java, C#): ")
        questions = self.question_bank.questions.get(topic, [])
        if not questions:
            print("No questions available for this topic.")
            return
        score = 0
        total_questions = len(questions)

        for question, correct_answer, *options in questions:
            print("\n" + question)
            for option in options:
                print(option)
            answer = input("Your answer (a/b/c/d): ").strip().lower()
            if answer in ['a', 'b', 'c', 'd']:
                if answer == correct_answer:
                    score += 10
            else:
                print("Invalid answer. Counting as 0.")

        print(f"\nYour total score: {score} out of {total_questions * 10}")

    def submit_feedback(self):
        if self.current_user.role != 'Employee':
            print("Access denied. Only employees can submit feedback.")
            return
        feedback = input("Please provide your feedback: ")
        print("Thank you for your feedback!")

    def download_question_bank(self):
        if self.current_user.role != 'Employee':
            print("Access denied. Only employees can download the question bank.")
            return
        print("Downloading question bank...")
        print("Question Bank Downloaded. Please Check the folder C:/User /Downloads.")

    def trainer_menu(self):
        while True:
            print("\nTrainer Menu:")
            print("1. Upload Curriculum")
            print("2. Generate Question Bank")
            print("3. Edit Question Bank")
            print("4. Exit")
            choice = input("Select an option: ")
            if choice == '1':
                self.upload_curriculum()
            elif choice == '2':
                self.generate_question_bank()
            elif choice == '3':
                self.edit_question_bank()
            elif choice == '4':
                break
            else:
                print("Invalid option. Please try again.")

    def upload_curriculum(self):
        curriculum = input("Enter the curriculum details: ")
        print(f"Curriculum uploaded: {curriculum}")

    def generate_question_bank(self):
        topic = input("Enter the topic for the question bank ( Python, Java, C#): ")
        if topic in self.question_bank.questions:
            questions = self.question_bank.questions[topic]
            print(f"Question bank generated for {topic}.")
            for question in questions:
                print(question[0])  # Print the question
        else:
            print("No questions available for this topic.")

    def edit_question_bank(self):
        topic = input("Enter the topic to edit the question bank (Python, Java, C#): ")
        if topic in self.question_bank.questions:
            print(f"Current questions for {topic}:")
            for idx, question in enumerate(self.question_bank.questions[topic]):
                print(f"{idx + 1}. {question[0]}")
            question_num = int(input("Enter the question number to edit: ")) - 1
            if 0 <= question_num < len(self.question_bank.questions[topic]):
                new_question = input("Enter the new question: ")
                # Here, you might want to ask for the options as well
                self.question_bank.questions[topic][question_num] = (new_question, 'a', 'a) Option1', 'b) Option2', 'c) Option3', 'd) Option4')
                print("Question updated successfully.")
            else:
                print("Invalid question number.")
        else:
            print("No questions available for this topic.")

    def admin_menu(self):
        employees = [
            "Alice",
            "Bob",
            "Charlie",
            "David",
            "Eve"
        ]
        skills = {
            'Java': ['Java FSD', 'Spring', 'Hibernate'],
            'Python': ['Django', 'Flask', 'Data Science'],
            'C#': ['.NET', 'ASP.NET', 'Entity Framework']
        }

        employee_data = {}
        for employee in employees:
            language = random.choice(list(skills.keys()))
            topic = random.choice(skills[language])
            employee_data[employee] = {
                'skills': {language: random.randint(1, 10), topic: random.randint(1, 10)},
                'performance': [random.randint(60, 100) for _ in range(24)]
            }

        while True:
            print("\nAdministrator Menu:")
            print("1. View Employee Analysis")
            print("2. Generate Performance Reports")
            print("3. Exit")
            choice = input("Select an option: ")
            if choice == '1':
                print("\nSelect an employee for analysis:")
                for idx, employee in enumerate(employees):
                    print(f"{idx + 1}. {employee}")
                emp_choice = int(input("Enter the number of the employee: ")) - 1

                if 0 <= emp_choice < len(employees):
                    selected_employee = employees[emp_choice]
                    data = employee_data[selected_employee]
                    self.display_employee_analysis(selected_employee, data)
                else:
                    print("Invalid employee selection.")
            elif choice == '2':
                print("\nSelect an employee for performance analysis:")
                for idx, employee in enumerate(employees):
                    print(f"{idx + 1}. {employee}")
                emp_choice = int(input("Enter the number of the employee: ")) - 1

                if 0 <= emp_choice < len(employees):
                    selected_employee = employees[emp_choice]
                    data = employee_data[selected_employee]
                    self.display_performance_analysis(selected_employee, data)
                else:
                    print("Invalid employee selection.")
            elif choice == '3':
                break
            else:
                print("Invalid option. Please try again.")

    def display_employee_analysis(self, employee, data):
        skills = data['skills']
        labels = list(skills.keys())
        sizes = list(skills.values())

        plt.figure(figsize=(8, 6))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title(f'Skill Distribution of {employee}')
        plt.axis('equal')
        plt.show()

    def display_performance_analysis(self, employee, data):
        months = [f'Month {i + 1}' for i in range(24)]
        performance = data['performance']

        plt.figure(figsize=(10, 5))
        plt.plot(months, performance, marker='o')
        plt.title(f'Performance Analysis of {employee}')
        plt.xlabel('Months')
        plt.ylabel('Performance Score')
        plt.xticks(rotation=45)
        plt.grid()
        plt.ylim(0, 110)
        plt.axhline(y=75, color='r', linestyle='--', label='Passing Score (75)')
        plt.legend()
        plt.tight_layout()
        plt.show()

    def employee_menu(self):
        while True:
            print("\nEmployee Menu:")

            print("1. Self Assessment")
            print("2. Submit Feedback")
            print("3. Download Question Bank")
            print("4. Exit")
            choice = input("Select an option: ")
            if choice == '1':
                self.self_assessment()
            elif choice == '2':
                self.submit_feedback()
            elif choice == '3':
                self.download_question_bank()
            elif choice == '4':
                break
            else:
                print("Invalid option. Please try again.")

    def run(self):
        while True:
            print("\nAutomated Question Builder Application")
            print("1. Register User")
            print("2. Exit")
            choice = input("Select an option: ")
            if choice == '1':
                username = input("Enter username: ")
                role = input("Enter role (Administrator/Trainer/Employee): ")
                self.register_user(username, role)
                if role == 'Administrator':
                    self.admin_menu()
                elif role == 'Trainer':
                    self.trainer_menu()
                elif role == 'Employee':
                    self.employee_menu()
                else:
                    print("Invalid role. Please try again.")
            elif choice == '2':
                print("Exiting application.")
                break
            else:
                print("Invalid option. Please try again.")

if __name__ == "__main__":
    app = AutomatedQuestionBuilderApp()
    app.run()
```
